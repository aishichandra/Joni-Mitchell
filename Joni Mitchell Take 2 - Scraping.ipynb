```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
def get_soundtrack_df(movies):
    
    soundtrack_list = []    
    for movie in movies.find_all("div",class_="filmo-category-section")[3].find_all('div', attrs={'class':["filmo-row odd","filmo-row even"]}):
    #     print(movie)
        name = movie.find('a').text

    #     print(name)
        res = re.findall(r'\(.*?\)', movie.text)


        if res[0].startswith('(writer') | res[0].startswith('(performer') | res[0].startswith('(lyrics') | res[0].startswith('(music'):
            category = 'Movie'

        else:
            category = res[0].replace('(', "").replace(')', "")

        year = ""

        year = movie.find('span', class_="year_column").text.replace('\n', "").strip()
    #     print(year.text)
        writer = ""
        performer = ""


        for attr in res:
            hello = ""
    #         print(attr)
            if attr.replace('(', "").replace(')', "").isnumeric():
                year = attr.replace('(', "").replace(')', "")
            if attr.startswith('(writer:'):
                writer = attr.replace('(', "").replace(')', "").split(":")[1].replace("- uncredited", "").strip()
            if attr.startswith('(lyrics:'):
                writer = attr.replace('(', "").replace(')', "").split(":")[1].replace("- uncredited", "").strip()
            if attr.startswith('(music:'):
                writer = attr.replace('(', "").replace(')', "").split(":")[1].replace("- uncredited", "").strip()
            if attr.startswith('(performer:'):
                performer = attr.replace('(', "").replace(')', "").split(":")[1].strip().replace("- uncredited", "").strip()
            if performer != writer:
                hello = "False"
    #         print(hello)
        dictionary = {'name': name, 'category': category, 'year': year, 'song': writer, 'performer': performer, 'bool': hello}
    #     print(dictionary)
    #     print("---------------------------------------------------------------")
        soundtrack_list.append(dictionary)
    return soundtrack_list
```

```python
response = requests.get('https://www.imdb.com/name/nm0001168/')
# response = requests.get('https://www.imdb.com/name/nm0593474/')

movies = BeautifulSoup(response.text)
```

```python
soundtrack_list = get_soundtrack_df(movies)

df = pd.DataFrame(soundtrack_list)
df[df.song.notnull()].head(60)
```
